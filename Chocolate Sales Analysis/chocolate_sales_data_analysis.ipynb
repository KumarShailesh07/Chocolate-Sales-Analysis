```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
import webbrowser
import os
```

```python
sales_df = pd.read_csv("Chocolate Sales.csv")
sales_df
```

```python
sales_df.tail()
```

```python
#pd.read_csv() : csv files
#pd.read_excel() : excel files
#pd.read_sql() : SQL Databases
#pd.read_json() : JSON Files

#df.isnull() : Missing values
#df.dropna() : Removes rows and columns that contain the missing values
#df.fillna() : Fills missing values

#df.duplicated() : Identifies duplicates
#df.drop_duplicates() : Removes duplicate rows
```

```python
#Data Cleaning
sales_df = sales_df.dropna(subset=["Boxes Shipped"])
for column in sales_df.columns :
    sales_df[column].fillna(sales_df[column].mode()[0],inplace=True)
sales_df.drop_duplicates(inplace=True)
```

```python
sales_df.dtypes
```

```python
sales_df
```

```python
sales_df.dtypes
```

```python
#Data Transformation
def remove_symbol(Amount):
    if '$' in Amount:
        return Amount.replace('$','')
    else:
        return np.nan  #Here nan means null value
sales_df['Amount'] = sales_df['Amount'].apply(remove_symbol)
```

```python
sales_df["Amount"] = sales_df["Amount"].astype(str).str.replace(',','').astype(int)
```

```python
sales_df
```

```python
sales_df.dtypes
```

```python
sales_df
```

```python
sales_df.dtypes
```

```python
sales_df['Log_Boxes_Shipped'] = np.log1p(sales_df['Boxes Shipped'])
sales_df['Log_Amount'] = np.log1p(sales_df['Amount'])
```

```python
sales_df
```

```python
# Add Revenue column
sales_df['Revenue'] = sales_df['Amount'] * sales_df['Boxes Shipped']
```

```python
sales_df
```

```python
#To change date,month,year in numeric form
#Here we change the column into datetime format
sales_df['Date'] = pd.to_datetime(sales_df['Date'],errors='coerce')
```

```python
#Now we extracting the year from that column
sales_df['Year'] = sales_df['Date'].dt.year
```

Creating Graph using plotly

```python
html_files_path="./"
if not os.path.exists(html_files_path):
    os.makedirs(html_files_path) #makedirs : make a directry means create a path
```

```python
plot_containers=""
```

```python
# Save each Plotly figure to an HTML file
def save_plot_as_html(fig, filename, insight):
    global plot_containers
    filepath = os.path.join(html_files_path, filename)
    html_content = pio.to_html(fig, full_html=False, include_plotlyjs='inline')
    # Append the plot and its insight to plot_containers
    plot_containers += f"""
    <div class="plot-container" id="{filename}" onclick="openPlot('{filename}')">
        <div class="plot">{html_content}</div>
        <div class="insights">{insight}</div>
    </div>
    """
    fig.write_html(filepath, full_html=False, include_plotlyjs='inline')
```

```python
plot_width=400
plot_height=300
plot_bg_color='black'
text_color='white'
title_font={'size':16}
axis_font={'size':12}
```

```python
sales_df
```

```python
#Figure 1
product_shipment_counts = sales_df.groupby('Product')['Boxes Shipped'].sum().nlargest(10)
fig1 = px.bar(
    x=product_shipment_counts.index,
    y=product_shipment_counts.values,
    labels={'x': 'Product', 'y': 'Total Boxes Shipped'},
    title='Top 10 Products Boxes Shipped',
    color=product_shipment_counts.index,
    color_discrete_sequence=px.colors.sequential.Plasma,
    width=400,
    height=300,  # Adjusted for better readability
)
fig1.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
    font_color='white',
    title_font={'size':16},
    xaxis=dict(title_font={'size':12}),
    yaxis=dict(title_font={'size':12}),
    margin=dict(l=10,r=10,t=30,b=10)
)
#fig1.update_traces(marker=dict(pattern=dict(line=dict(color='white',width=1))))
save_plot_as_html(fig1,"product_shipment_counts1.html","The top shipped products include Fruit & Nut Bars, Eclairs, Manuka Honey Choco, and 50% Dark Bites, showing their strong market demand. Some specialty items, like Organic Choco Syrup, have significant shipments in specific regions, suggesting niche appeal. High shipment volumes often correlate with popular consumer preferences and strong sales channels.")
            
```
